```python
from selenium import *
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service = Service(ChromeDriverManager().install()))
```

```python
import pandas as pd
import numpy as np
import time

temp = pd.read_csv('nowon_rest.csv',sep=',', encoding='CP949')
temp = temp[['지번주소', '사업장명', '업태구분명']]
temp.columns = ['address', 'name', 'category']
temp = temp.drop_duplicates(['name'], keep='first')
```

```python
from tqdm.notebook import tqdm

temp['kakao_keyword'] = "공릉동 " + temp['name']
temp['kakaomap_url'] = ''

for i, keyword in enumerate(tqdm(temp['kakao_keyword'].tolist())):
    print(i, f"/ {temp.shape[0]} 행", keyword)
    try:
        kakao_map_search_url = f"https://map.kakao.com/?q={keyword}"
        driver.get(kakao_map_search_url)
        time.sleep(1.5)
        #a = driver.find_element(By.CSS_SELECTOR, '#info\.search\.place\.list > li:nth-child(1) > div.info_item > div.contact.clickArea > a.moreview').get_attribute('href')
        #print(a)
        temp.iloc[i,-1] = driver.find_element(By.CSS_SELECTOR, '#info\.search\.place\.list > li:nth-child(1) > div.info_item > div.contact.clickArea > a.moreview').get_attribute('href')
    except Exception as e1:
        if "li:nth-child(1)" in str(e1):
            try:
                temp.iloc[i,-1] = driver.find_element(By.CSS_SELECTOR,"#info\.search\.place\.list > li > div.info_item > div.contact.clickArea > a.moreview").get_attribute('href')
                time.sleep(1.5)
            except Exception as e2:
                print(e2)
                temp.iloc[i,-1] = np.nan
                time.sleep(1.5)
        else:
            pass

temp = temp.dropna()
temp.to_csv("crawled.csv", index=False, encoding="utf-8-sig")
```

```python
crawled = pd.read_csv('crawled.csv',sep=',')

crawled['kakao_name'] = ''
crawled['kakao_category'] = ''
crawled['kakao_star'] = ''
crawled['kakao_star_num'] = ''

review = pd.read_csv('reviews.csv',sep=',')

k=0

for i, url in enumerate(tqdm(crawled['kakaomap_url'])):
    driver.get(url)
    time.sleep(2.5)
    review_text = ""
    
    try:
        kakao_map_name = driver.find_element(By.CSS_SELECTOR, '#mArticle > div.cont_essential > div:nth-child(1) > div.place_details > div > h2').text
        kakao_map_category = driver.find_element(By.CSS_SELECTOR, '#mArticle > div.cont_essential > div:nth-child(1) > div.place_details > div > div > span.txt_location').text
        kakao_map_star = driver.find_element(By.CSS_SELECTOR, '#mArticle > div.cont_essential > div:nth-child(1) > div.place_details > div > div > a:nth-child(3) > span.color_b.inactive').text
        kakao_map_star_review_num = "0"
    except Exception as e1:
        try:
            if "inactive" in str(e1):
                kakao_map_star = driver.find_element(By.CSS_SELECTOR, "#mArticle > div.cont_essential > div:nth-child(1) > div.place_details > div > div > a:nth-child(3) > span.color_b").text
                kakao_map_star_review_num = driver.find_element(By.CSS_SELECTOR, '#mArticle > div.cont_evaluation > strong > span').text
            else:
                print(f"{i}행 {crawled.loc[i,'name']} 평점 갯수 크롤링 에러")
        except:
            kakao_map_star = "0"
            kakao_map_star_review_num = "0"
    
    crawled.loc[i,'kakao_name'] = kakao_map_name
    crawled.loc[i,'kakao_category'] = kakao_map_category
    crawled.loc[i,'kakao_star'] = kakao_map_star
    crawled.loc[i,'kakao_star_num'] = kakao_map_star_review_num
            
    j=0
    try:
        while j<3:
            try:
                j+=1
                review_text = driver.find_element(By.CSS_SELECTOR, f"#mArticle > div.cont_evaluation > div.evaluation_review > ul > li:nth-child({j}) > div.comment_info > p > span").text
                score = driver.find_element(By.CSS_SELECTOR, f"#mArticle > div.cont_evaluation > div.evaluation_review > ul > li:nth-child({j}) > div.star_info > div > span > span").get_attribute('style')
                #print(score)
                if len(score) < 11:
                    score = int(score[6:8])/20
                elif len(score) < 12:
                    score = int(score[7:9])/20
                else:
                    score = int(score[7:10])/20
                #print(score)
                review.loc[k,'kakao_name'] = kakao_map_name
                review.loc[k,'review_star'] = score
                review.loc[k,'kakao_review'] = review_text
                #print(review.loc[k,'review_star'])
                if j==3:
                    break
            except Exception as e1:
                #print(e1)
                if "li:nth-child(1)" in str(e1):
                    print(f"{i}행 {crawled.loc[i,'kakao_name']} 리뷰 0개")
                    review_text = "empty"
                    review.loc[k,'kakao_name'] = kakao_map_name
                    review.loc[k,'review_star'] = score
                    review.loc[k,'kakao_review'] = review_text
                    break
                else:
                    print(f"{i}행 {crawled.loc[i,'kakao_name']} 리뷰 {j - 1}개")  
                    review.loc[k,'kakao_name'] = kakao_map_name
                    review.loc[k,'review_star'] = score
                    review.loc[k,'kakao_review'] = review_text
                    break
            k+=1
    except Exception as e2:
        print(e2)
        print(f"{i}행 {crawled.loc[i,'kakao_name']} 문제가 발생 - code 2")
        
crawled.to_csv('crawled_test.csv', index=True, encoding="utf-8-sig")
review.to_csv('review_test.csv', index=True, encoding="utf-8-sig")

driver.quit()
```
